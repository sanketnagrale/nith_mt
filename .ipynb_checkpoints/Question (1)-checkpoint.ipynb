# Question

Write the code for a sequence of 50 number belonging to a sine wave, predict the 51st number in the series, build a model and train the model with 2 RNN and 3 LSTM layers for 25% dropout.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import LSTM, SimpleRNN, Dense, Embedding
from tensorflow.keras.models import Sequential

import numpy
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
x = np.arange(0, 10, 0.1)
y = np.sin(x)
```

```python
def windowed_dataset(series, window_size, batch_size, shuffle_buffer):
  dataset = tf.data.Dataset.from_tensor_slices(series)
  dataset = dataset.window(window_size + 1, shift=1, drop_remainder=True)
  dataset = dataset.flat_map(lambda window: window.batch(window_size + 1))
  dataset = dataset.shuffle(shuffle_buffer).map(lambda window: (window[:-1], window[-1]))
  dataset = dataset.batch(batch_size).prefetch(1)
  return dataset
```

```python
window_size = 10
split_time = 80
batch_size = 16
shuffle_buffer_size = 100
time_train = x[:split_time]
x_train = y[:split_time]
time_valid = x[split_time:]
x_valid = y[split_time:]
```

```python
dataset = windowed_dataset(x_train, window_size, batch_size, shuffle_buffer_size)
print(dataset)

model = tf.keras.models.Sequential([
  tf.keras.layers.Lambda(lambda x: tf.expand_dims(x, axis=-1),
                      input_shape=[None]),
  tf.keras.layers.SimpleRNN(40, return_sequences=True),
  tf.keras.layers.SimpleRNN(40),
  tf.keras.layers.Dense(1),
  tf.keras.layers.Lambda(lambda x: x * 100.0)
])
optimizer = tf.keras.optimizers.SGD(lr=5e-5, momentum=0.9)
model.compile(loss=tf.keras.losses.Huber(),
              optimizer=optimizer,
              metrics=["mae"])
history = model.fit(dataset,epochs=400)
```

```python
forecast = []

for time in range(len(y)-window_size):
  forecast.append(model.predict(y[time:time+window_size][np.newaxis]))


results = np.array(forecast)[:, 0, 0]
```

```python
def plot_series(time, series, format="-", start=0, end=None):
    plt.plot(time[start:end], series[start:end], format)
    plt.xlabel("Time")
    plt.ylabel("Value")
    plt.grid(True)

plt.figure(figsize=(10, 6))

plot_series(x, y)
plot_series(x[10:], results)
```
